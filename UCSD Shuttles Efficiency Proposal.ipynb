# Background and Prior Work

The average cost for CNG (Compressed Natural Gas) is two dollars and fifty cents per gallon, which is what the UCSD shuttles use. With this in mind, we can use data collected from UCSD shuttles every time they fuel to determine how much gas is used per day for each route. About 5.3 percent of Californians commute to work by public transit (which is not a lot in the grand scheme of things). Most of our group members work at UCSD shuttles so it will be easier for us to gether data for this project and analyze it in the context of the these real world statistics. Based on our real world experiences and "passenger logs", which log how many people get on our shuttles, we can determine if the ratio of passengers to the student population exceeds 5.3 percent or if it is lower.

# Hypothesis

Our routes are not cost efficient or environmentally friendly because there are less passengers than would be sufficient to maintain certain routes. Some buses have the same coverage or overlap stops/routes and combining them would save resources. There are some times in the day where buses run for two hours (four loops) with less than five passengers overall.

# Data

The ideal data set would have gallons of fuel per day for each route (and the money to gallon ratio), how many buses per route, hours of service, and passengers per loop (and at what times). We will be using observations from approximately ten different routes using data from 2018, early 2019, and perhaps some data from this quarter if possible. This data would be collected from UCSD shuttle administrative spreadsheets as well as "passenger logs" that we will put in allowed buses. The drivers would record the passenger intake and the gallons of gas because it is already a requirement for their jobs. This data will be stored in a CSV file for easy manipulation and visualization creation.

# Ethics & Privacy

The ethics and privacy portion of this project is ideal because the passenger logs only document the quantity of passengers and not any attributes specific to the individuals. The benefits of this project outweigh the risks as most data is collected with complete anonymity. The only part that is not anonymous is which drivers fuel and log the gallons of CNG put into the bus, but this is necessary to monitor driver performance for their jobs, which makes the collection of data ethical.
